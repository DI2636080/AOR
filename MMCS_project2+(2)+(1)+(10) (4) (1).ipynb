```python
num_time_slots = 34*1
num_movies = 100
max_time = 700 # time in seconds
num_channels=3
num_genre = 19

T = list(range(num_time_slots))  # example of T (you can adjust the size)
M = list(range(num_movies))  # example of M (you can adjust the size)
T1 = list(range(33,num_time_slots-1,34)) # 11.19
T_ad = {0:list(range(63)),1:list(range(63)),2:list(range(62))}
C = list(range(num_channels))  
G = list(range(num_channels))
prob = xp.problem(name='myproblem')
#xp.init('/Applications/FICO Xpress/xpressmp/bin/xpauth.xpr')
#xp.init('C:/xpressmp/bin/xpauth.xpr')
#variables
x = {(movie_database.loc[m,'title'], t): xp.var(vartype=xp.binary, name='x_{0}_{1}'.format(movie_database.loc[m,'title'], t))
     for t in T for m in M}
y = {movie_database.loc[m,'title']: xp.var(vartype=xp.binary, name='y_{0}'.format(movie_database.loc[m,'title'])) for m in M}
s = {movie_database.loc[m,'title']: xp.var(vartype=xp.continuous, name='s_{0}'.format(movie_database.loc[m,'title'])) for m in M}
e = {movie_database.loc[m,'title']: xp.var(vartype=xp.continuous, name='e_{0}'.format(movie_database.loc[m,'title'])) for m in M}
N = 1000000
z = {(m, c, t): xp.var(vartype=xp.binary, name=f'z_{m}_{c}_{t}')for m in M for c in C for t in T_ad[c]}
# 11.23  flag to mark whether at least one movie genre is same as ads genre
f = {(movie_database.loc[m, 'title'], t, c, g): xp.var(vartype=xp.binary, name='f_{0}_{1}_{2}_{3}'.format(movie_database.loc[m, 'title'], t, c, g)) 
    for c in C for t in T_ad[c] for m in M for g in G}

#z = {(movie_database.loc[m, 'title'], t, c): xp.var(vartype=xp.binary, name='x_{0}_{1}_{2}'.format(movie_database.loc[m, 'title'], t, c)) 
#     for c in C for t in T for m in M}

prob.addVariable(x,y,s,e,z)

channal_0_time_slots = [32, 63, 95, 127, 160, 192, 225]
channal_1_time_slots = [33, 63, 94, 125, 157, 187, 219]
channal_2_time_slots = [32, 62, 94, 126, 159, 191, 223]
T2 = list(range(channal_0_time_slots[6]))
T3 = list(range(channal_1_time_slots[6]))
T4 = list(range(channal_2_time_slots[6]))
n=2
#constraint
# assum ads budget
#total_budget = xp.Sum(ad_slot_price_0.loc[c, 'ad_slot_price'] + ad_slot_price_1.loc[c, 'ad_slot_price'] + ad_slot_price_2.loc[c, 'ad_slot_price'] * z[(c, m, t)]
#                      for c in C for m in M for t in T)
#ad_movie= [xp.Sum(z[movie_database.loc[m,'title'],c,t] for m in M for t in T for c in C) <= y[movie_database.loc[m,'title']]]
#ad_min_in_day = [t * z[movie_database.loc[m, 'title'], c, t] <= s[movie_database.loc[m, 'title']] * 30
#    for m in M for t in T for c in C]


by_day = [x[movie_database.loc[m,'title'], t] <= 1 - x[movie_database.loc[m,'title'], t+1] for t in T1 for m in M] # 11.19
assignment = [xp.Sum(x[movie_database.loc[m,'title'], t] for m in M) == 1 for t in T]
full_duration = [xp.Sum(x[movie_database.loc[m,'title'], t] for t in T) ==y[movie_database.loc[m,'title']]*movie_database.loc[m,'runtime_with_ads']/30 for m in M]
start_time = [s[movie_database.loc[m,'title']] <= t*x[movie_database.loc[m,'title'], t] + (1-x[movie_database.loc[m,'title'], t])*num_time_slots for t in T for m in M]
end_time = [e[movie_database.loc[m,'title']] >= (t+1)*x[movie_database.loc[m,'title'], t]  for t in T for m in M]
run_time = [e[movie_database.loc[m,'title']] - s[movie_database.loc[m,'title']] == y[movie_database.loc[m,'title']]*movie_database.loc[m,'runtime_with_ads']/30 for m in M]
advert_const = [xp.Sum(z.get((movie_database.loc[m, 'title'], c, t), 0) for m in M) <= 1 for c in C for t in T_ad[c]]
# only show ads if our channel stream the movie
# if the ads genre is the same as the movie
#advert_const4= [y[movie_database.loc[m,'title']]*mg_list[m,g] >= f[(movie_database.loc[m,'title'], t,c,g)] for c in C for t in T_ad[c] for m in M for g in G]
#advert_const5 = [
 # z.get((movie_database.loc[m, 'title'], c, t), 0) * adg_list[c, t, g] >= f.get((movie_database.loc[m, 'title'], t, c, g), 0)
 # for c in C for t in T_ad[c] for m in M for g in G]
#advert_const6= [xp.Sum(f[(movie_database.loc[m,'title'], t,c,g)] for g in G) >= z.get((movie_database.loc[m,'title'], c, t), 0) for c in C for t in T_ad[c] for m in M]

# advertise constraint   11.23
# ads show before movie
advert_const2 = [5 * ad_t_list[c][t] * z.get((movie_database.loc[m,'title'], c, t), 0) <= 30 * s[movie_database.loc[m,'title']] 
                 for m in M for c in C for t in T_ad[c]] 

# only show ads if our channel stream the movie
advert_const3 = [
    xp.Sum(z.get((movie_database.loc[m,'title'], c, t), 0) for c in C for t in T_ad[c]) 
    <= num_time_slots * xp.Sum(x[movie_database.loc[m,'title'], t] for t in T)
    for m in M
]

prob.addConstraint(assignment,full_duration,start_time,end_time,run_time,by_day,advert_const2,advert_const,advert_const3)#,advert_const6)

#objective function
weighted_view_count =xp.Sum(x[movie_database.loc[m, 'title'], t] * (
      df.loc[t, 'children_baseline_view_count'] * movie_database.loc[m, 'children_scaled_popularity'] + 
       df.loc[t, 'adults_baseline_view_count'] * movie_database.loc[m, 'adults_scaled_popularity'] + 
      df.loc[t, 'retirees_baseline_view_count'] * movie_database.loc[m, 'retirees_scaled_popularity']
   )  for t in T for m in M)* N

weighted_conversion_rate_count =xp.Sum(
    (
        xp.Sum((df_0.loc[t, 'Maximum Conversion rate'] * df_sch_0.loc[t, 'children_true_view_count_1'] +
         df_0.loc[t, 'Maximum Conversion rate'] * df_sch_0.loc[t, 'adults_true_view_count_1'] +
         df_0.loc[t, 'Maximum Conversion rate'] * df_sch_0.loc[t, 'retirees_true_view_count_1']) * z[(m, 0, t)] for t in T_ad[0])+
        xp.Sum((df_1.loc[t, 'Maximum'] * df_sch_1.loc[t, 'children_true_view_count_1'] +
         df_1.loc[t, 'Maximum'] * df_sch_1.loc[t, 'adults_true_view_count_1'] +
         df_1.loc[t, 'Maximum'] * df_sch_1.loc[t, 'retirees_true_view_count_1']) * z[(m, 1, t)] for t in T_ad[1])+
        xp.Sum((df_2.loc[t, 'Maximum'] * df_sch_2.loc[t, 'children_true_view_count_1'] +
         df_2.loc[t, 'Maximum'] * df_sch_2.loc[t, 'adults_true_view_count_1'] +
         df_2.loc[t, 'Maximum'] * df_sch_2.loc[t, 'retirees_true_view_count_1']) * z[(m, 2, t)] for t in T_ad[2])
    ) * N for m in M)

#Minimize advertsing cost=total budget
objective = weighted_view_count + weighted_conversion_rate_count
#objective2= objective2= xp.Sum(movie_database.loc[m,'license fee']*y[movie_database.loc[m,'title']]+xp.Sum(
# df_sch_0.loc[c,'ad_slot_price'] for t in T2[channal_0_time_slots[n-2]:channal_1_time_slots[n-1]]+df_sch_1.loc[c,'ad_slot_price'] for t in T3[channal_1_time_slots[n-2]:channal_1_time_slots[n-1]]
#    +df_sch_2.loc[c,'ad_slot_price'] for t in T4[channal_2_time_slots[n-2]:channal_2_time_slots[n-1]])*z[(c, m, t)] for c in C for m in M)
# objective2 = xp.Sum(
#     movie_database.loc[m, 'license fee'] * y[movie_database.loc[m, 'title']] + 
#     xp.Sum(
#         df_sch_0.loc[t, 'ad_slot_price']*z[(0, m, t)] for t in T
#    ) + 
#     xp.Sum(
#         df_sch_1.loc[t, 'ad_slot_price']*z[(1, m, t)] for t in T
#     ) +
#     xp.Sum(
#         df_sch_2.loc[t, 'ad_slot_price']*z[(2, m, t)] for t in T
#     )for m in M)

objective2 = xp.Sum(
    movie_database.loc[m, 'license fee'] * y[movie_database.loc[m, 'title']] + 
    xp.Sum(
        df_sch_0.loc[t, 'ad_slot_price']*z[(m, 0, t)] for t in T2[channal_0_time_slots[n-2]:channal_0_time_slots[n-1]]
   ) + 
    xp.Sum(
        df_sch_1.loc[t, 'ad_slot_price']*z[(m, 1, t)] for t in T3[channal_1_time_slots[n-2]:channal_1_time_slots[n-1]]
    ) +
    xp.Sum(
        df_sch_2.loc[t, 'ad_slot_price']*z[(m, 2, t)] for t in T4[channal_2_time_slots[n-2]:channal_2_time_slots[n-1]]
    )for m in M)
# Combine both objective functions into a single objective (example: you want to minimize objective2 and maximize objective)
combined_objective = 0.9*objective - 0.1*objective2  # Or you can scale them if you need a balance between them

# Set the objective function to maximize the combined objective
prob.setObjective(objective, sense=xp.maximize)  # Or xp.maximize if you prefer to minimize the result


```

```python
import pandas as pd

# 假设 prob.getSolution() 返回了决策变量的值
# 例如，这里假设 s_sol1, s_sol2, s_sol3 是 xpress 表达式，可以通过 prob.getSolution() 获取求解的值。

# 创建一个字典来存储每个 (movie, time_slot) 的结果
s_sol1_dict = {}
s_sol2_dict = {}
s_sol3_dict = {}

# 假设 T 是时间槽的集合，M 是电影的集合
for t in T:
    for m in M:
        # 获取 s_sol1, s_sol2, s_sol3 对应的结果
        s_sol1_dict[(movie_database.loc[m, 'title'], t)] = prob.getSolution(df.loc[t, 'children_baseline_view_count'] * movie_database.loc[m, 'children_scaled_popularity']) * N
        s_sol2_dict[(movie_database.loc[m, 'title'], t)] = prob.getSolution(df.loc[t, 'adults_baseline_view_count'] * movie_database.loc[m, 'adults_scaled_popularity']) * N
        s_sol3_dict[(movie_database.loc[m, 'title'], t)] = prob.getSolution(df.loc[t, 'retirees_baseline_view_count'] * movie_database.loc[m, 'retirees_scaled_popularity']) * N

# 将字典转换为 DataFrame 格式
data1 = [(movie, time_slot, value) for (movie, time_slot), value in s_sol1_dict.items()]
data2 = [(movie, time_slot, value) for (movie, time_slot), value in s_sol2_dict.items()]
data3 = [(movie, time_slot, value) for (movie, time_slot), value in s_sol3_dict.items()]

# 创建 DataFrame
df_s_sol1 = pd.DataFrame(data1, columns=['Movie', 'Time Slot', 'Children Viewership'])
df_s_sol2 = pd.DataFrame(data2, columns=['Movie', 'Time Slot', 'Adults Viewership'])
df_s_sol3 = pd.DataFrame(data3, columns=['Movie', 'Time Slot', 'Retirees Viewership'])

# 合并所有的 DataFrame
df_combined = pd.merge(df_s_sol1, df_s_sol2, on=['Movie', 'Time Slot'], how='outer')
df_combined = pd.merge(df_combined, df_s_sol3, on=['Movie', 'Time Slot'], how='outer')

# 保存为 CSV 文件
df_combined.to_csv('viewership_results.csv', index=False)

print("CSV 文件已保存!")
```
